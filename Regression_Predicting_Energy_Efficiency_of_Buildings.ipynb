```python
df = df.drop(columns='lights')
df.head()
```

```python
scaler = MinMaxScaler()
normalised = scaler.fit_transform(df)
```

```python
X = np.array(df["T2"]).reshape(-1,1)
y = np.array(df["T6"]).reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)
```

```python
Linear_model = LinearRegression()
Linear_model.fit(X, y)
y_predict = Linear_model.predict(X_test)
```

```python
r2_score = r2_score(y_test, y_predict)
round(r2_score, 2)
```

```python
mae = mean_absolute_error(y_test, y_predict)
round(mae, 2)
```

```python
Linear_model.coef_
```

```python
rss = np.sum(np.square(y_test-y_predict))
round(rss, 2)
```

```python
rmse = sqrt(mean_squared_error(y_test, y_predict))
round(rmse, 2)
```

```python
model2 = Ridge(alpha=0.4)
model2.fit(X, y)
y_predict2 = model2.predict(X_test)
```

```python
rmse = sqrt(mean_squared_error(y_test, y_predict2))
round(rmse, 2)
```

```python
model3 = Lasso(alpha=0.001)
model3.fit(X, y)
y_predict3 = model3.predict(X_test)
```

```python
rmse = sqrt(mean_squared_error(y_test, y_predict3))
round(rmse, 3)
```
